### Variables Description 

From the above decription we observe that:

The Minimum 'Price' is 0 which is not possible.

The Maximum 'minimum_nights' is 1250 which is also highly unlikely.

````python
df[df['price']==0]
#Rows which have 'price'=0
````

````python
# defining a function for price correction
def price_correct(x):
  if x==0:
    x=100
  return x
````

````python
# Using the price_correction function to change the 'price'
df['price']=df['price'].apply(price_correct)
````

````python
df[df['price']==0]
````

````python
#correct values of 'minimum_nights' to 365 which are greater than 365
df['minimum_nights']=df['minimum_nights'].apply(lambda x: 365 if x>365 else x)
````

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

````python
# Chart - 1 visualization code
max_hostid=df['host_id'].value_counts().head(10)
max_hostid
````

#Considering 10 highest host_id listings.The maximum host_id listing is 327.

````python
#creating a dataframe with max_hostid data
df_hostid=pd.DataFrame(data=max_hostid)
df_hostid.reset_index(inplace=True)
df_hostid.rename(columns={'index':'Host_ID', 'host_id':'Listings_Count'}, inplace=True)
df_hostid
````

````python
df_hostid['Host_ID']=df_hostid['Host_ID'].apply(str)
````

````python
plt.figure(figsize=(20,10))
plt.bar(df_hostid['Host_ID'],df_hostid['Listings_Count'], color ='maroon',)
plt.xlabel("Host ID")
plt.ylabel("Listings Count")
plt.title("Hosts with most listing count in NYC")
plt.show()
````

##### 1. Why did you pick the specific chart?

Above graph shows top 10 host with highest number of listings

##### 2. What is/are the insight(s) found from the chart?

Highest no of listing is 372 which is by host '219517861'
